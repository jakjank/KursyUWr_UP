```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

data = pd.read_csv("./iris/iris.data", header=None)
setosas = data[data[4] == "Iris-setosa"]

x = [0] + list(np.arange(2.5, math.ceil(np.max(setosas[1]))+0.5, 0.5))
counts = {}

for i in range(len(x) - 1):
    lower = x[i]
    upper = x[i + 1]
    count = len(setosas[(setosas[1] >= lower) & (setosas[1] < upper)])
    counts[f"{lower}-{upper}"] = count

fig, ax = plt.subplots()

ax.bar(counts.keys(), counts.values())

plt.title("# of setosas with certain sepal width")
plt.xlabel("sepal width (cm)")
plt.ylabel("count")

plt.show()
```

Input variables:

1 - age (numeric)

2 - job : type of job (categorical: "admin.","unknown","unemployed","management","housemaid","entrepreneur","student",
                                   "blue-collar","self-employed","retired","technician","services") 
                                   
3 - marital : marital status (categorical: "married","divorced","single"; note: "divorced" means divorced or widowed)

4 - education (categorical: "unknown","secondary","primary","tertiary")

5 - default: has credit in default? (binary: "yes","no")

6 - balance: average yearly balance, in euros (numeric) 

7 - housing: has housing loan? (binary: "yes","no")

8 - loan: has personal loan? (binary: "yes","no")

9 - contact: contact communication type (categorical: "unknown","telephone","cellular") 

10 - day: last contact day of the month (numeric)

11 - month: last contact month of year (categorical: "jan", "feb", "mar", ..., "nov", "dec")

12 - duration: last contact duration, in seconds (numeric)

13 - campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)

14 - pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric, -1 means client was not previously contacted)

15 - previous: number of contacts performed before this campaign and for this client (numeric)

16 - poutcome: outcome of the previous marketing campaign (categorical: "unknown","other","failure","success")

Output variable (desired target):

17 - y - has the client subscribed a term deposit? (binary: "yes","no")

```python
data = pd.read_csv("./bank_marketing/bank/bank-full.csv", sep=";")

def get_positive_percentage(time):
    data_t = data[data["duration"] > time]
    no_samples = len(data_t) // 5
    # print(f"{no_samples} samples for duration > {time}")

    if no_samples == 0:
        return 0
    
    samples = data_t.sample(n=no_samples, replace=False)

    pos = len(samples[samples['y'] == 'yes'])
    
    return pos / no_samples * 100

duration = range(0, 4080, 400)
positive_percentage = []

for t in duration:
    positive_percentage.append(get_positive_percentage(t))

fig, ax = plt.subplots()
ax.plot(duration, positive_percentage, )
plt.xticks(duration)
plt.yticks(range(0,100,10))
plt.xlabel("last contact duration")
plt.ylabel("% of positive samples")
plt.show()
```

```python
data = pd.read_csv("./bank_marketing/bank/bank-full.csv", sep=";")
data1 = data.sample(frac=0.5)
data2 = data.drop(data1.index)

balance = range(0, np.max(data1["balance"]), 1000)
positive_percentage = []

def get_percentage(t, data, positive=True, above=True):
    if above:
        data_t = data[data["balance"] > t]
    else:
        data_t = data[data["balance"] < t]
    if positive:
        x = len(data_t[data_t['y'] == 'yes'])
    else:
        x = len(data_t[data_t['y'] == 'no'])
    # print(t, " - ", len(data_t))
    return x / len(data_t) * 100

for t in balance:
    positive_percentage.append(get_percentage(t, data1))
    
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(balance, positive_percentage, )
plt.xticks(range(0, np.max(data["balance"]), 10000))
plt.yticks(range(0,100,10))
plt.xlabel("yearly balance")
plt.ylabel("% of positive samples")
plt.show()
```

### Grouped bar plot for different values of balance threshold

```python
# decided to cut off 70_000 - max range bc there are some 4 ppl there
balance_thresholds = range(0, 70000, 10000)
groups = {}

# def get_percentage(t, data, positive=True, above=True):

for bt in balance_thresholds:
    positive_above1 = get_percentage(bt, data1, True, True)
    positive_above2 = get_percentage(bt, data2, True, True)
    negative_above1 = get_percentage(bt, data1, False, True)
    negative_above2 = get_percentage(bt, data2, False, True)
    positive_below1 = get_percentage(bt, data1, True, False)
    positive_below2 = get_percentage(bt, data2, True, False)
    groups[bt] = [positive_above1, positive_above2, negative_above1, negative_above2, positive_below1, positive_below2]

fig, ax = plt.subplots(figsize=(14, 5))
width = 0.12

br1 = np.arange(len(balance_thresholds))
br2 = [x+width for x in br1]
br3 = [x+width for x in br2]
br4 = [x+width for x in br3]
br5 = [x+width for x in br4]
br6 = [x+width for x in br5]


plt.bar(br1, [gr[0] for gr in groups.values()], width=width, label='positive % above thr. (data1)')
plt.bar(br2, [gr[1] for gr in groups.values()], width=width, label='positive % above thr. (data2)')
plt.bar(br3, [gr[2] for gr in groups.values()], width=width, label='negative % above thr. (data1)')
plt.bar(br4, [gr[3] for gr in groups.values()], width=width, label='negative % above thr. (data2)')
plt.bar(br5, [gr[4] for gr in groups.values()], width=width, label='positive % below thr. (data1)')
plt.bar(br6, [gr[5] for gr in groups.values()], width=width, label='positive % below thr. (data2)')
   

ax.set_xlabel("Threshold")
ax.set_ylabel("%")
ax.set_xticks(np.arange(len(balance_thresholds)) + (6 - 1) * width / 2)
ax.set_xticklabels(balance_thresholds)
plt.legend()

plt.show()

    
    
```
